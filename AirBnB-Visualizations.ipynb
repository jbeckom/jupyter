```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
listings = pd.read_csv('Airbnb_NYC_2019.csv')
```

```python
listings
```

```python
#show listings by price
plt.hist(listings['price'])
plt.xlabel('price (USD)')
plt.show()
```

```python
#limit price to 1100
plt.hist(listings['price'], bins = np.arange(0,1100,40))
plt.xlabel('price (USD)')
plt.show()
```

```python
#show number of reviews by price
plt.scatter(x=listings['price'],y=listings['number_of_reviews'])
plt.xlabel('price (USD)')
plt.ylabel('number of reviews')
plt.show()
```

```python
#limit price to $1100
plt.scatter(x=listings['price'],y=listings['number_of_reviews'])
plt.xlabel('price (USD)')
plt.ylabel('number of reviews')
plt.xlim(0,1100)
plt.show()
```

```python
#limit price to $1100, reduce size of plots
plt.scatter(x=listings['price'],y=listings['number_of_reviews'],s=5)
plt.xlabel('price (USD)')
plt.ylabel('number of reviews')
plt.xlim(0,1100)
plt.show()
```
